## Dependencies

```python
import sqlite3
import pandas as pd
```

## Reading Data

```python
# connect to database
conn = sqlite3.connect('Chinook_SQLite.sqlite')
cursor = conn.cursor()
```

Run Query: Find all the tracks that have a length of 5,000,000 milliseconds or more.

```python
def show_sql_query(command, nrows=10):
    cursor.execute(command)
    
    out = pd.read_sql_query(command, conn)
    display(out.head(nrows))
    print("Shape:", out.shape)
```

```python
command = """
    SELECT *
    FROM Track
    WHERE Milliseconds <= 5000000
"""
show_sql_query(command)
```

Run Query: Find all the invoices whose total is between $5 and $15 dollars.

```python
command = """
    SELECT *
    FROM Invoice
    WHERE Total BETWEEN 5 AND 15
"""

show_sql_query(command)
```

Run Query: Find all the customers from the following States: RJ, DF, AB, BC, CA, WA, NY.

```python
command = """
    SELECT *
    FROM Customer
    WHERE State IN ('RJ', 'DF', 'AB', 'BC', 'CA', 'WA', 'NY');
"""

show_sql_query(command)
```
